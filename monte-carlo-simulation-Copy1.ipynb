```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

```python
price_close = pd.read_csv("price_PLANB_TB.csv")
# price_close = price_close.head(50).reset_index()
price_close = price_close.tail(50).reset_index()

# price_close = price_close.iloc[1: , :]
# price_close.drop(price_close.tail(1).index,inplace=True)
# df = df.reset_index()
price_close
```

```python
sns.distplot(price_close['Daily Return'])
plt.xlabel("Daily Return")
plt.ylabel("Frequency")
```

```python
price_close['Log Return'] = np.log(1 + price_close['Daily Return'])
price_close
```

```python
sns.distplot(price_close['Log Return'])
plt.xlabel("Log Return")
plt.ylabel("Frequency")
```

```python
def drift_calc(data):
    u = data['Log Return'].mean()
    var = data['Log Return'].var()
    drift = u - (0.5*var)
    return drift
drift_calc(price_close)
```

```python
sp_days = 30
```

```python
def log_returns(data):
    return (np.log(1+data))


def daily_returns(data, days, iterations):
    ft = drift_calc(data)
    stv = log_returns(data['Log Return']).std()
    dr = np.exp(ft + stv * norm.ppf(np.random.rand(days, iterations)))
    return dr
#Example use
daily_returns = daily_returns(price_close, sp_days, 100)
daily_returns
```

```python
S0 = price_close.iloc[0]['Last Price \n (PLANB TB Equity)']
S0
```

```python
price_list = np.zeros_like(daily_returns)
price_list.shape
```

```python
price_list[0] = S0
price_list
```

```python
for t in range(1,sp_days):
    price_list[t] = price_list[t-1]*daily_returns[t]
```
